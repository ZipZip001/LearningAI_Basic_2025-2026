```python
# Initialize
openai = OpenAI()
MODEL = 'gpt-4.1-mini'
```

```python
system_message = "Bạn là một người giải thích cho tôi cách mà AI hoạt động"
```

```python
# Chatbox
def chat(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    response = openai.chat.completions.create(
        model = MODEL,
        messages = messages
    )
    return response.choices[0].message.content
```

```python
gr.ChatInterface(fn=chat).launch()
```

```python
#Using steam to impove UX/UI
def chatStrem(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    stream = openai.chat.completions.create(
        model = MODEL,
        messages = messages,
        stream = True
    )
    response = "" # Why ?
    for chuck in stream:
        response += chuck.choices[0].delta.content or ''
        yield response 
```

```python
gr.ChatInterface(fn=chatStrem).launch()
```
